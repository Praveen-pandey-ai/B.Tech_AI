# Video

```python
import cv2

# Make sure the path is correct and the video file exists
video = cv2.VideoCapture('C:\\Users\\257383\\Downloads\\__ Nobita X Yeh Badal Ki Chadar Yeh Taro Ke Achal ____ Doremon Whatsapp Status .mp4')

# Check if the video opened successfully
while True:
    ret, frame = video.read() # ye function 2 values ko return karta hai 1st is boolean and second is image
    frame = cv2.resize(frame, (550, 550))
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY) # To convert Black and white video into grayscale
    cv2.imshow('video',frame)
    cv2.imshow('Gray',gray)

    k =  cv2.waitKey(50)# here values of wait key will decide the speed of frames(video), 0 value means static frame
    if k == ord('q'): # To save the video
        break
video.release()
cv2.destroyAllWindows()

```
